```python
import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
# Create the category dictionary
def create_category_dict(data_path):
    
    categories = os.listdir(data_path)
    
    labels = [i for i in range(len(categories))]
    
    category_dict = {}
    for i in range(len(categories)):
        category_dict[categories[i]] = labels[i]
        
    return category_dict, categories
```

```python
data_path = 'dataset'

category_dict, categories = create_category_dict(data_path)

print(category_dict, categories)
```

```python
# test code to see the all images
for category in categories:
    imgs_path = os.path.join(data_path, category)
    img_names = os.listdir(imgs_path)
        
    for img_name in img_names:
        img_path = os.path.join(imgs_path, img_name)
        img = cv2.imread(img_path, 0)
        cv2.imshow('LIVE', img)
        k = cv2.waitKey(100)
        if k==27:
            break
cv2.destroyAllWindows()
```

```python
# Create data & target data sets
def create_data_target(data, target):
    for category in categories:
        imgs_path = os.path.join(data_path, category)
        img_names = os.listdir(imgs_path)
        print(imgs_path, '---------------------------------')
        print(img_names)
        
        for img_name in img_names:
            img_path = os.path.join(imgs_path, img_name)
            img = cv2.imread(img_path, 0)
            img = cv2.resize(img, (8,8))
            data.append(img)
            target.append(category_dict[category])
    
    print('')
    print("Completed.")
    print(data)
    print(target)

    return data, target
```

```python
data = []
target = []

data, target = create_data_target(data, target)
```

```python
# print the shape of the data and target
print(f'data: {np.array((data)).shape}, target: {np.array(target).shape}')
```
